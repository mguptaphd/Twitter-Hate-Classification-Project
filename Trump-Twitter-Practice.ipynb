```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
trump_twitter_url = requests.get('https://twitter.com/realDonaldTrump').text
```

```python
soup = BeautifulSoup(trump_twitter_url,'lxml')
#print(soup) 
soup
```

```python
tweets = soup.find_all("p")
tweets
#df = pd.read_html(tweets)
```

```python
tweets = soup.find_all("p",{"class":"TweetTextSize"})
tweets
```

```python
tweet_list =[]
for tweet in tweets:
    tweet_list.append(tweet.text)  
#by using .text you are pulling out only the text (no html tags)
#append to a new list 
```

```python
df = pd.DataFrame(tweet_list)
df.columns = ['TrumpTweet']
df
```

```python
df.to_csv('data/trump_tweet.csv')
```
